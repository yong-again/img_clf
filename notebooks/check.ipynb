```python
train_df = pd.read_csv(r'C:\works\dacon\img_clf\data\train_csv\train_file_list.csv')
label_dict = train_df[['folder_name', 'label']].drop_duplicates().set_index('label').to_dict()['folder_name']
rst2['label'] = rst2['prediction'].map(lambda x: label_dict.get(x, 'unknown'))
sample_df = pd.read_csv(r'C:\works\dacon\img_clf\data\sample_submission.csv')
```

```python
# Prepare empty DataFrame with same columns as sample_df
submission_df = pd.DataFrame(0, index=range(len(rst2)), columns=sample_df.columns)

# Assign ID from image_path (remove extension)
submission_df["ID"] = rst2["image_path"].str.replace(".jpg", "", regex=False)

# For each row, set the corresponding label column to 1
for idx, label in enumerate(rst2["label"]):
    if label in submission_df.columns:
        submission_df.at[idx, label] = 1
```

```python
submission_df.to_csv(r'../data/result/submission2.csv', index=False)
```
